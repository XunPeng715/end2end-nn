GO
- the same as <start> on the picture below - the first token which is fed to the decoder along with the thoughvector in order to start generating tokens of the answer

EOS 
- "end of sentence" - the same as <end> on the picture below - as soon as decoder generates this token we consider the answer to be complete (you can't use usual punctuation marks for this purpose cause their meaning can be different)

UNK 
- "unknown token" - is used to replace the rare words that did not fit in your vocabulary. So your sentence My name is guotong1988 will be translated into My name is _unk_.

PAD 
- your GPU (or CPU at worst) processes your training data in batches and all the sequences in your batch should have the same length. If the max length of your sequence is 8, your sentence My name is guotong1988 will be padded from either side to fit this length: My name is guotong1988 _pad_ _pad_ _pad_ _pad_

```python
from collections import Counter
count.extend(Counter(words).most_common())
```

```python
count
```

```python
count[1][0]
```

```python
word2idx = {}
            
if len(word2idx) == 0:
    word2idx['<eos>'] = 0
    
for word, _ in count:
    if word not in word2idx:
        word2idx[word] = len(word2idx )
```

```python
word2idx
```

```python
data = list()
for line in lines:
    for word in line.split():
        index = word2idx[word]
        data.append(index)
    data.append(word2idx['<eos>'])
```
